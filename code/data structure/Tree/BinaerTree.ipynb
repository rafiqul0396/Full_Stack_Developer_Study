```python
#Maximum Depth of Binary Tree
#Given a binary tree, find its maximum depth.
#The maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.
#Note: A leaf is a node with no children.
#Example:
#Given binary tree [3,9,20,null,null,15,7],
#    3
#   / \
#  9  20
#    /  \
#   15   7
#return its depth = 3.
#Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
class Solution:
    def maxDepth(self, root: TreeNode) -> int:
        if root is None:
            return 0
        else:
            left_height=self.maxDepth(root.left)
            right_height=self.maxDepth(root.right)
            return max(left_height,right_height)+1
#driver code
if __name__ == "__main__":
    root=TreeNode(3)
    root.left=TreeNode(9)
    root.right=TreeNode(20)
    root.right.left=TreeNode(15)
    root.right.right=TreeNode(7)
    print(Solution().maxDepth(root))
```

```python
#Minimum Depth of Binary Tree
#Given a binary tree, find its minimum depth.
#The minimum depth is the number of nodes along the shortest path from the root node down to the nearest leaf node.
#Note: A leaf is a node with no children.
#Example:
#Given binary tree [3,9,20,null,null,15,7],
#    3
#   / \
#  9  20
#    /  \
#   15   7
#return its minimum depth = 2.
#Definition for a binary tree node.

def minimum_depth(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return 1
    if root.left is None:
        return minimum_depth(root.right)+1
    if root.right is None:
        return minimum_depth(root.left)+1
    return min(minimum_depth(root.left),minimum_depth(root.right))+1
#driver code
if __name__ == "__main__":
    root=TreeNode(3)
    root.left=TreeNode(9)
    root.right=TreeNode(20)
    root.right.left=TreeNode(15)
    root.right.right=TreeNode(7)
    print(minimum_depth(root))
```

```python
#Symmetric Tree
#Given a binary tree, check whether it is a mirror of itself (ie, symmetric around its center).
#For example, this binary tree [1,2,2,3,4,4,3] is symmetric:
#    1
#   / \
#  2   2
# / \ / \
#3  4 4  3
#But the following [1,2,2,null,3,null,3] is not:
#    1
#   / \
#  2   2
#   \   \
#   3    3
#Note:
#Bonus points if you could solve it both recursively and iteratively.
#Definition for a binary tree node.


def isSymmetric(root):
    if root is None:
        return True
    def isMirror(left,right):
        if left is None and right is None:
            return True
        if left is None or right is None:
            return False
        return (left.val==right.val) and isMirror(left.left,right.right) and isMirror(left.right,right.left)
    return isMirror(root.left,root.right)

#driver code
if __name__ == "__main__":
    root=TreeNode(1)
    root.left=TreeNode(2)
    root.right=TreeNode(2)
    root.left.left=TreeNode(3)
    root.left.right=TreeNode(4)
    root.right.left=TreeNode(4)
    root.right.right=TreeNode(3)
    print(isSymmetric(root))
```

```python
#same tree
#Given two binary trees, write a function to check if they are the same or not.
#Two binary trees are considered the same if they are structurally identical and the nodes have the same value.
#Example 1:
#Input:     1         1
#          / \       / \
#         2   3     2   3
#        [1,2,3],   [1,2,3]
#Output: true

#Example 2:
#Input:     1         1
#          /           \
#         2             2
#        [1,2],     [1,null,2]
#Output: false


#Example 3:
#Input:     1         1
#          / \       / \
#         2   1     1   2
#        [1,2,1],   [1,1,2]
#Output: false
#Definition for a binary tree node.


def isSameTree(p, q):
    if p is None and q is None:
        return True
    if p is None or q is None:
        return False
    return (p.val==q.val) and isSameTree(p.left,q.left) and isSameTree(p.right,q.right)

#driver code
if __name__ == "__main__":
    p=TreeNode(1)
    p.left=TreeNode(2)
    p.right=TreeNode(1)
    q=TreeNode(1)
    q.left=TreeNode(1)
    q.right=TreeNode(2)
    print(isSameTree(p,q))
    
```

```python
#Invert Binary Tree
#Invert a binary tree.
#Example:
#Input:
#     4
#   /   \
#  2     7
# / \   / \
#1   3 6   9
#Output:
#     4
#   /   \
#  7     2
# / \   / \
#9   6 3   1
#Trivia:
#This problem was inspired by this original tweet by Max Howell:

#Google: 90% of our engineers use the software you wrote (Homebrew),
#but you canâ€™t invert a binary tree on a whiteboard so f*** off.


def invertTree(root):
    if root is None:
        return None
    root.left,root.right=root.right,root.left
    invertTree(root.left)
    invertTree(root.right)
    return root

#driver code
if __name__ == "__main__":
    root=TreeNode(4)
    root.left=TreeNode(2)
    root.right=TreeNode(7)
    root.left.left=TreeNode(1)
    root.left.right=TreeNode(3)
    root.right.left=TreeNode(6)
    root.right.right=TreeNode(9)
    print(invertTree(root))
    
```

```python
#Binary Tree Level Order Traversal
#Given a binary tree, return the level order traversal of its nodes' values. (ie, from left to right, level by level).
#For example:
#Given binary tree [3,9,20,null,null,15,7],
#    3
#   / \
#  9  20
#    /  \
#   15   7
#return its level order traversal as:

#[
#  [3],
#  [9,20],
#  [15,7]
#]
#Definition for a binary tree node.



def levelOrder(root):
    if root is None:
        return []
    queue=[root]
    result=[]
    while queue:
        level=[]
        for i in range(len(queue)):
            node=queue.pop(0)
            level.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        result.append(level)
    return result

#driver code
if __name__ == "__main__":
    root=TreeNode(3)
    root.left=TreeNode(9)
    root.right=TreeNode(20)
    root.right.left=TreeNode(15)
    root.right.right=TreeNode(7)
    print(levelOrder(root))
```

```python
#Binary Tree Zigzag Level Order Traversal
#Given a binary tree, return the zigzag level order traversal of its nodes' values. (ie, from left to right, then right to left for the next level and alternate between).
#For example:
#Given binary tree [3,9,20,null,null,15,7],
#    3
#   / \
#  9  20
#    /  \
#   15   7
#return its zigzag level order traversal as:
#[
#  [3],
#  [20,9],
#  [15,7]
#]
#Definition for a binary tree node.



def zigzagLevelOrder(root):
    if root is None:
        return []
    queue=[root]
    result=[]
    flag=0
    while queue:
        level=[]
        for i in range(len(queue)):
            node=queue.pop(0)
            level.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        if flag==0:
            result.append(level)
            flag=1
        else:
            result.append(level[::-1])
            flag=0
    return result


#driver code
if __name__ == "__main__":
    root=TreeNode(3)
    root.left=TreeNode(9)
    root.right=TreeNode(20)
    root.right.left=TreeNode(15)
    root.right.right=TreeNode(7)
    print(zigzagLevelOrder(root))
```

```python
#Binary Tree Maximum Path Sum
#Given a non-empty binary tree, find the maximum path sum.
#For this problem, a path is defined as any node sequence from some starting node to any node in the tree along the parent-child connections. The path must contain at least one node and does not need to go through the root.
#Example 1:
#Input: [1,2,3]
#       1
#      / \
#     2   3
#Output: 6
#Example 2:
#Input: [-10,9,20,null,null,15,7]
#   -10
#   / \
#  9  20
#    /  \
#   15   7
#Output: 42
#Definition for a binary tree node.

def maxPathSum(root):
    if root is None:
        return 0
    left=maxPathSum(root.left)
    right=maxPathSum(root.right)
    return max(left+right+root.val,left,right)
def findPathSum(root):
    if root is None:
        return 0
    left=findPathSum(root.left)
    right=findPathSum(root.right)
    return max(0,root.val+max(left,right))

def pathsumII(root,sum):
    if root is None:
        return []
    if root.left is None and root.right is None and root.val==sum:
        return [[root.val]]
    left=pathsumII(root.left,sum-root.val)
    right=pathsumII(root.right,sum-root.val)
    return [[root.val]+i for i in left+right]
def pathsumIII(root,sum):
    if root is None:
        return 0
    return pathsumII(root,sum)+pathsumIII(root.left,sum)+pathsumIII(root.right,sum)

def sumRootLeaf(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return root.val
    left=sumRootLeaf(root.left)
    right=sumRootLeaf(root.right)
    return left*10+right+root.val

def maxRootLeaf(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return root.val
    left=maxRootLeaf(root.left)
    right=maxRootLeaf(root.right)
    return max(left,right)+root.val

def minRootLeaf(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return root.val
    left=minRootLeaf(root.left)
    right=minRootLeaf(root.right)
    return min(left,right)+root.val

def sumNumbers(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return root.val
    left=sumNumbers(root.left)
    right=sumNumbers(root.right)
    return left*10+right+root.val*2


def maxLeaf(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return root.val
    left=maxLeaf(root.left)
    right=maxLeaf(root.right)
    return max(left,right)
def leafNode(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return 1
    left=leafNode(root.left)
    right=leafNode(root.right)
    return left+right

def sumleafNode(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return root.val
    left=sumleafNode(root.left)
    right=sumleafNode(root.right)
    return left+right
def maxandminleaf(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return root.val
    left=maxandminleaf(root.left)
    right=maxandminleaf(root.right)
    return [max(left,right),min(left,right)]

def search(root,target):
    if root is None:
        return False
    if root.val==target:
        return True
    left=search(root.left,target)
    right=search(root.right,target)
    return left or right

def Largest(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return root.val
    left=secondLargest(root.left)
    right=secondLargest(root.right)
    return max(left,right)
def MinNode(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return root.val
    left=MinNode(root.left)
    right=MinNode(root.right)
    return min(left,right)
def secondLargest(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return root.val
    left=Largest(root.left)
    right=Largest(root.right)
    if left>right:
        return max(left,MinNode(root.right))
    else:
        return max(right,MinNode(root.left))
def MinNode(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return root.val
    left=MinNode(root.left)
    right=MinNode(root.right)
    return min(left,right)

def distinctNode(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return 1
    left=distinctNode(root.left)
    right=distinctNode(root.right)
    return left+right


#driver code

if __name__ == "__main__":
    root=TreeNode(1)
    root.left=TreeNode(2)
    root.right=TreeNode(3)
    #root.left.left=TreeNode(2)
    print(findPathSum(root))
    print(maxPathSum(root))
    print(pathsumII(root,3))
    #print(pathsumIII(root,3))
    print(sumRootLeaf(root))
    print(maxRootLeaf(root))
    print(minRootLeaf(root))
    print(sumNumbers(root))
    print(maxLeaf(root))
    print(leafNode(root))
    print(sumleafNode(root))
    print(maxandminleaf(root))
    print(search(root,3))
    print(secondLargest(root))
    print(secondLargest(root))
    print(distinctNode(root))

```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def verticalOrder(root: TreeNode):
    # Base case
    if not root:
        return []

    # Dictionary to store nodes at each vertical distance
    nodes_dict = {}

    # DFS function to traverse the tree and add nodes to dictionary
    def dfs(node, col, row):
        if node:
            nodes_dict[col] = nodes_dict.get(col, []) + [(row, node.val)]
            dfs(node.left, col-1, row+1)
            dfs(node.right, col+1, row+1)

    # Start DFS traversal from root
    dfs(root, 0, 0)

    # Sort the keys (vertical distances) of the dictionary
    cols = sorted(nodes_dict.keys())

    # Create result list of nodes in each column
    res = []
    for col in cols:
        nodes = [val[1] for val in sorted(nodes_dict[col])]
        res.append(nodes)

    return res


#driver code
if __name__ == "__main__":
    root=TreeNode(1)
    root.left=TreeNode(2)
    root.right=TreeNode(3)
    root.left.left=TreeNode(4)
    root.left.right=TreeNode(5)
    root.right.left=TreeNode(6)
    root.right.right=TreeNode(7)
    print(verticalOrder(root))
```
